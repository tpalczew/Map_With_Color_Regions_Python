```python
import matplotlib.pyplot as plt
import matplotlib.cm
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.colors import Normalize
```

```python
# group data county wise 

# Group the data state-wise
county_group = df.groupby(by = "SchoolCounty")
print(county_group.size())
```

```python
number_of_outlets_per_county = []
for key , group in county_group:
    total_county_out = 0
    for row in group.iterrows():
        total_county_out += row[1][3]
        
    number_of_outlets_per_county.append((key,total_county_out))

print(number_of_outlets_per_county)
```

```python
#   Bounding Box to get region coordinates
# http://boundingbox.klokantech.com
#
# westlimit=-79.89; southlimit=39.88; eastlimit=-70.99; northlimit=44.97
#

#  Shapes
#  https://www.census.gov/geo/maps-data/data/cbf/cbf_counties.html
#
```

```python
m = Basemap(resolution='f', # c, l, i, h, f or None
            projection='merc',
            lat_0=54.5, lon_0=-4.36,
            llcrnrlon=-79.89, llcrnrlat=39.88, urcrnrlon=-70.99, urcrnrlat=44.97)
```

```python
# draw coastlines
```

```python
fig, ax = plt.subplots(figsize=(20,20))

#m.drawmapboundary(fill_color='#46bcec')

m.fillcontinents(color='coral',lake_color='aqua')
m.readshapefile("/Users/tpalczew/cb_2017_us_county_20m/cb_2017_us_county_20m","cb_2017_us_county_20m")
m.drawcoastlines()
```

```python

outlets_byc = np.zeros(len(m.cb_2017_us_county_20m_info))
outlets_byc = outlets_byc.tolist()

i = 0
for county_info in m.cb_2017_us_county_20m_info:
    # Get the state in uppercase, as our csv file has state names in upper case
    county = (county_info['NAME'])
    outlets = 0
    for x in number_of_outlets_per_county:
        if x[0] == county:
            outlets = x[1]
            break
        
    outlets_byc[i] = outlets
    i += 1

print(outlets_byc)
```

```python
df_poly = pd.DataFrame({
        'shapes': [Polygon(np.array(shape), True) for shape in m.cb_2017_us_county_20m],
        'area': [area['NAME'] for area in m.cb_2017_us_county_20m_info],
        'outlets_byc' : outlets_byc
        })

df_poly['outlets_byc'].replace(0, np.NaN)
df_poly = df_poly.dropna(how="all") 


df_poly
```

```python
fig, ax = plt.subplots()

m = Basemap(resolution='f', # c, l, i, h, f or None
            projection='merc',
            lat_0=54.5, lon_0=-4.36,
            llcrnrlon=-79.89, llcrnrlat=39.88, urcrnrlon=-70.99, urcrnrlat=44.97)

#m.drawmapboundary(fill_color='#46bcec')

m.drawmapboundary(fill_color='white')



m.fillcontinents(color='#f2f2f2',lake_color='aqua')
m.readshapefile("/Users/tpalczew/cb_2017_us_county_20m/cb_2017_us_county_20m","cb_2017_us_county_20m")
m.drawcoastlines()

shapes = [Polygon(np.array(shape), True) for shape in m.cb_2017_us_county_20m]
cmap = plt.get_cmap('Oranges')   
pc = PatchCollection(shapes, zorder=2)
norm = Normalize()

pc.set_facecolor(cmap(norm(df_poly['outlets_byc'].fillna(0).values)))
ax.add_collection(pc)


mapper = matplotlib.cm.ScalarMappable(cmap=cmap)
mapper.set_array(outlets_byc)
plt.colorbar(mapper, shrink=0.4)

ax.set_title("Number of School's water outlets per county")



plt.rcParams['figure.figsize'] = (20,20)
plt.rcParams.update({'font.size': 20})

plt.savefig("num_of_school_water_outl.png")
plt.show()
```
